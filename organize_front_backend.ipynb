# Organize Frontend and Backend Folders
This notebook moves the frontend and backend projects from `artifacts/` into top-level `frontend/` and `backend/` folders, then removes the unused `artifacts/mockup-sandbox` folder.


```python
import shutil
from pathlib import Path

root = Path(r"c:\Users\USER\Downloads\LocalHelper")
frontend_src = root / "artifacts" / "erp"
backend_src = root / "artifacts" / "api-server"
mockup_src = root / "artifacts" / "mockup-sandbox"
frontend_dst = root / "frontend"
backend_dst = root / "backend"

print('root exists:', root.exists())
print('frontend source exists:', frontend_src.exists())
print('backend source exists:', backend_src.exists())
print('mockup source exists:', mockup_src.exists())

if frontend_src.exists():
    if frontend_dst.exists():
        print('frontend destination already exists')
    else:
        shutil.move(str(frontend_src), str(frontend_dst))
        print('moved frontend')
else:
    print('frontend source missing')

if backend_src.exists():
    if backend_dst.exists():
        print('backend destination already exists')
    else:
        shutil.move(str(backend_src), str(backend_dst))
        print('moved backend')
else:
    print('backend source missing')

if mockup_src.exists():
    shutil.rmtree(mockup_src)
    print('removed mockup-sandbox')
else:
    print('mockup-sandbox not present')

artifacts_dir = root / 'artifacts'
if artifacts_dir.exists() and not any(artifacts_dir.iterdir()):
    artifacts_dir.rmdir()
    print('removed empty artifacts folder')
else:
    print('artifacts directory still present')
```
